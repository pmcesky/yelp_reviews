```python
# clean 'text' column
review['text'] = review['text'].apply(lambda x: x.removeprefix('b'))
review['text'] = review['text'].apply(lambda x: x.removeprefix('\''))
review['text'] = review['text'].apply(lambda x: x.removeprefix('"'))
review['text'] = review['text'].apply(lambda x: x.removesuffix('\''))
review['text'] = review['text'].apply(lambda x: x.removesuffix('"'))
```

```python
# clean other str columns
for col in ['date','business_id','user_id','review_id']:
    review[col] = review[col].apply(lambda x: x.removeprefix('b'))
    review[col] = review[col].apply(lambda x: x.removeprefix('\''))
    review[col] = review[col].apply(lambda x: x.removeprefix('"'))
    review[col] = review[col].apply(lambda x: x.removesuffix('\''))
    review[col] = review[col].apply(lambda x: x.removesuffix('"'))
```

```python
review
```

For stars in review, if stars>=4, we say its sentiment is positive, if stars<=2, we say its sentiment is negative.
And we can create the 'sentiment' column for the review dataset.

```python
review['sentiment'] = None
review.loc[review['stars']>=3.9, 'sentiment'] = 1
review.loc[review['stars']<=2.1, 'sentiment'] = 0
review
```

```python
review.to_csv('./yelp_academic_dataset_review_first_151K.csv', index=False)
```

We remove stars=3.0 as this is neither positive or negative sentiment. And we save the remaining part of the review.

```python
review_valid_sentiment = review.dropna(ignore_index=True)
review_valid_sentiment
```

We keep the first 130K entries of sentiments.

```python
review_valid_sentiment = review_valid_sentiment.head(130000)
review_valid_sentiment
```

```python
review_valid_sentiment.to_csv('./yelp_review_first_130K_with_sentiment.csv', index=False)
```
